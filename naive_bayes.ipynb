```python
dataset["daily_change"] = dataset["Adj Close"].pct_change()
```

```python
dataset["target"] = np.where(dataset["daily_change"] > 0, 1, 0)
dataset = dataset.iloc[1:]
dataset
```

```python
x = dataset[['Open','High','Low','Close','Adj Close','Volume','daily_change']]
x
```

```python
y = dataset['target']
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 0)  
```

```python
# Veri özellikleri (features) ölçeklendirilerek feature scaling işlemi yapılır.
from sklearn.preprocessing import StandardScaler  
sc = StandardScaler()  
x_train = sc.fit_transform(x_train)  
x_test = sc.transform(x_test)  
```

```python
from sklearn.naive_bayes import GaussianNB  
classifier = GaussianNB()  
classifier.fit(x_train, y_train)  
```

```python
y_predict = classifier.predict(x_test)  
y_predict
```

```python
from sklearn import metrics
print("Accuracy score %.3f" %metrics.accuracy_score(y_test, y_predict))
```

```python
mse = mean_squared_error(y_test, y_predict)
print("Mean Squared Error: %.2f" % mse)
rmse = np.sqrt(mse)
print("Root Mean Squared Error (RMSE): %.2f" % rmse)
```

```python
plt.figure(figsize=(15, 5))
# Gerçek değerler
plt.plot(y_test.values, color='blue', label='Gerçek Değer')

# Tahmin edilen değerler
plt.plot(y_predict, color='orange', label='Tahmin')

plt.xlabel('Örnekler',fontsize=15)
plt.ylabel('Fiyat Yönü',fontsize=15)
plt.title('Gerçek Değer ve Tahmin',fontsize=15)
plt.legend()
plt.show()
```
