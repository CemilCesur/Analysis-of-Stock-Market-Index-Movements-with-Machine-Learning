```python
# Set the stock symbol and download historical data
stock_symbol = selected_stock
data = yf.download(stock_symbol, period='1y')
data_ = data.head()
values = data.values
data
```

```python
# plot her kolon için
groups = [0, 1, 2, 3, 4, 5]
i = 1


pyplot.figure(figsize=(16,8))
for group in groups:
    pyplot.subplot(len(groups), 1, i)
    pyplot.plot(values[:, group])
    pyplot.title(data.columns[group], y=0.5, loc='right')
    i += 1
pyplot.show()
```

```python
#single graph according to closure
pyplot.figure(figsize=(16,8))
pyplot.title('Fiyat Kapanış Grafiği ('+selected_stock+')',fontsize=15)
pyplot.plot(data['Close'])
pyplot.xlabel('Tarih',fontsize=15)
pyplot.ylabel('Kapanış Fiyatı ',fontsize=15)
pyplot.show()
```

```python
# data hazırlığı
data['Date'] = data.index
data = data[['Date', 'Close']]
data.reset_index(drop=True, inplace=True)
data
```

```python
# Split the data into training and testing sets
train_data_len = int(0.7 * len(data))
train_data = data[:train_data_len]
test_data = data[train_data_len:]
```

```python
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_train_data = scaler.fit_transform(train_data['Close'].values.reshape(-1, 1))
```

```python
X_train = []
y_train = []

for i in range(60, len(scaled_train_data)):
    X_train.append(scaled_train_data[i-60:i, 0])
    y_train.append(scaled_train_data[i, 0])

X_train, y_train = np.array(X_train), np.array(y_train)
```

```python
# Train the Neural Network model
model = MLPRegressor(hidden_layer_sizes=(100, 100), activation='relu', random_state=42)
model.fit(X_train, y_train)
```

```python
inputs = data[len(data) - len(test_data) - 60:].values
inputs = inputs[:, 1]  # 'Close' değerlerini al
inputs = inputs.reshape(-1, 1).astype('float32')  # float32'ye dönüştür
inputs = scaler.transform(inputs)
inputs
```

```python
X_test = []
for i in range(60, inputs.shape[0]):
    X_test.append(inputs[i-60:i, 0])

X_test = np.array(X_test)
X_test
```

```python
predicted_stock_price = model.predict(X_test)
predicted_stock_price = scaler.inverse_transform(predicted_stock_price.reshape(-1, 1))
predicted_stock_price
```

```python
# Plot the results
plt.figure(figsize=(20, 10))
plt.plot(data['Date'], data['Close'], color='gray', label='Historical Price')
plt.plot(test_data['Date'], test_data['Close'], color='red', label='Actual Test Price')
plt.plot(test_data['Date'], predicted_stock_price, color='blue', label='Predicted Price')
plt.title(selected_stock+' Stock Price Prediction',fontsize=15)
plt.xlabel('Date',fontsize=15)
plt.ylabel('Closing Price',fontsize=15)
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
# y_test ve y_predict değişkenlerini tanımlama
y_test = test_data['Close']
y_predict = predicted_stock_price
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, y_predict)
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_predict)

def calculate_mape(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

# Verileri tek boyuta dönüştürme
y_test = y_test.ravel()
y_predict = y_predict.ravel()

# MAPE hesaplama
mape = calculate_mape(y_test, y_predict)

# RMSE hesaplama
rmse = np.sqrt(mse)

print("Mean Absolute Error (MAE): %.2f" % mae)
print("Mean Squared Error (MSE): %.2f" % mse)
print("Root Mean Squared Error (RMSE): %.2f" % rmse)
print("Mean Absolute Percentage Error (MAPE): %.2f" % mape)
```
